## 1. Data Preprocessing

```python
# Check data's Null values
data.isnull().any().any()
```

ÏúÑÏùò Í≤∞Í≥ºÎ•º ÌÜµÌï¥ Îç∞Ïù¥ÌÑ∞Ïóê Í≤∞Ï∏°Í∞íÏù¥ ÏûàÎã§Îäî Í≤ÉÏùÑ ÌôïÏù∏ÌñàÍ≥†, NullÍ∞íÏù¥ 900Í∞úÍ∞Ä ÎÑòÎäî ColumnÏùÄ ÏÇ≠Ï†úÌïòÍ≥† ÎÇòÎ®∏ÏßÄ ColumnÏùò NullÍ∞íÏùÄ 0ÏúºÎ°ú ÎåÄÏ≤¥ÌïòÎäî Ï†ÑÏ≤òÎ¶¨ Î∞©Î≤ïÏùÑ ÌÉùÌïúÎã§.

```python
# Í∞Å ColumnÎ≥ÑÎ°ú Í≤∞Ï∏°Í∞íÏùò Ïàò ÌôïÏù∏
d=data.isnull().sum()
d
```

```python
# NaNÍ∞íÏù¥ 900Í∞úÍ∞Ä ÎÑòÎäî Í≤ΩÏö∞ jÎ∞∞Ïó¥Ïóê Ï†ÄÏû•
j = []

for i in d.keys():
    if(d[i]>900):
        print(i,d[i])
        j.append(i)
```

```python
# Get rid of 'Time' column because it's meaningless
data=data.drop(['Time'],axis=1)

# DataÏóêÏÑú Í≤∞Ï∏°Í∞í 900Í∞ú ÎÑòÎäî j Column ÏÇ≠Ï†ú
data.drop(j, axis=1, inplace=True)
print(data.shape)
data
```

```python
# ÎÇ®ÏùÄ NaNÍ∞íÏùÄ 0ÏúºÎ°ú ÎåÄÏ≤¥
data.replace(np.nan, 0 , inplace=True)
data
```

```python
# Check data's Null values
data.isnull().any().any()
```

### Data Scaling
- 1. StandardScaler
- 2. PCA Ï∞®ÏõêÏ∂ïÏÜå

```python
# Train/Test Dataset splitÏùÑ ÏúÑÌï¥ DatasetÏùÑ Data/TargetÍ∞íÏúºÎ°ú Î∂ÑÎ¶¨
X = data.iloc[:, 1:-1]
y = data.iloc[:, -1]
```

(1) StandardScaler: Í∞Å featureÏùò ÌèâÍ∑†ÏùÑ 0, Î∂ÑÏÇ∞ÏùÑ 1Î°ú Î≥ÄÍ≤Ω, Î™®Îì† featureÎì§Ïù¥ Í∞ôÏùÄ ScaleÏùÑ Í∞ñÍ≤åÎêúÎã§.

```python
# Î®ºÏ†Ä Îç∞Ïù¥ÌÑ∞Î•º ÌïôÏäµÏö©Í≥º ÌÖåÏä§Ìä∏Ïö©ÏúºÎ°ú Î∂ÑÌï†
# test_test_splitÏùò random_stateÎ•º Ï†úÏô∏Ìïú Îã§Î•∏ ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Îäî defaultÎ°ú ÎëêÍ∏∞
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50)
```

#### ‚Äª ScalerÎ•º ÏÇ¨Ïö©ÌïòÍ∏∞ Ï†ÑÏóê Ï£ºÏùòÌï† Ï†ê
scalerÎäî fitÍ≥º transform Îß§ÏÑúÎìúÎ•º ÏßÄÎãàÍ≥† ÏûàÎã§. fit Îß§ÏÑúÎìúÎ°ú Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôòÏùÑ ÌïôÏäµÌïòÍ≥†, transform Îß§ÏÑúÎìúÎ°ú Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞Ïùò Ïä§ÏºÄÏùºÏùÑ Ï°∞Ï†ïÌïúÎã§.
Ïù¥Îïå! fit Îß§ÏÑúÎìúÎäî ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ÏóêÎßå Ï†ÅÏö©Ìï¥ÏïºÌïúÎã§. Í∑∏ ÌõÑ transform Îß§ÏÑúÎìúÎ•º ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ÏôÄ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Ïóê Ï†ÅÏö©ÌïúÎã§.

```python
from sklearn.preprocessing import StandardScaler 
sc= StandardScaler()
X_sc = sc.fit_transform(X)
X_train_sc = sc.fit_transform(X_train)
print('Features Min value before Scale : \n {}'. format(X_train.min(axis=0)))
print('Features Max value before Scale : \n {}'. format(X_train.max(axis=0)))
print('Features Min value after Scale : \n {}'. format(X_train_sc.min(axis=0)))
print('Features Max value after Scale : \n {}'. format(X_train_sc.max(axis=0)))
X_test_sc = sc.transform(X_test)
```

(2) PCAÏ∞®ÏõêÏ∂ïÏÜå

- ÎßéÏùÄ featureÎ°ú Íµ¨ÏÑ±Îêú Îã§Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏Ïùò Ï∞®ÏõêÏùÑ Ï∂ïÏÜåÌï¥ ÏÉàÎ°úÏö¥ Ï∞®ÏõêÏùò Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏Î•º ÏÉùÏÑ±ÌïòÎäî Í≤É
-  ÏùºÎ∞òÏ†ÅÏúºÎ°ú Ï∞®ÏõêÏù¥ Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù, Ï¶â featureÍ∞Ä ÎßéÏïÑÏßàÏàòÎ°ù ÏòàÏ∏° Ïã†Î¢∞ÎèÑÍ∞Ä Îñ®Ïñ¥ÏßÄÍ≥†, Í≥ºÏ†ÅÌï©(overfitting)Ïù¥ Î∞úÏÉùÌïòÍ≥†, Í∞úÎ≥Ñ featureÍ∞ÑÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÎÜíÏùÑ Í∞ÄÎä•ÏÑ±Ïù¥ ÏûàÏùå

```python
from sklearn.decomposition import PCA
# Initializing PCA and fitting
# Choose minimum number of PCA features that 85% variance is retained to avoid overfitting
pca = PCA(0.85)
X_sc_pca = pca.fit_transform(X_sc) # StandardScalerÏóêÏÑú ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞Ïóê Ï†ÅÏö©Ìïú Í≤ÉÏ≤òÎüº fitÍ≥º transform Îëò Îã§Î•º Ï†ÅÏö©Ìï¥Ï§òÏïº ÌïòÎÑ§~~??
```

```python
# Transform train and test datasets
X_train_pca = pca.transform(X_train_sc)
X_test_pca = pca.transform(X_test_sc)

print('PCAÏ∞®Ïõê Ï∂ïÏÜå Ïù¥Ï†Ñ X_train:', X_train_sc.shape)
print('PCAÏ∞®Ïõê Ï∂ïÏÜå Ïù¥Ï†Ñ X_test:', X_test_sc.shape)
print('PCAÏ∞®Ïõê Ï∂ïÏÜå Ïù¥ÌõÑ X_train:', X_train_pca.shape)
print('PCAÏ∞®Ïõê Ï∂ïÏÜå Ïù¥ÌõÑ X_test:', X_test_pca.shape)
```

PCAÏ∞®Ïõê Ï∂ïÏÜåÎ°ú FeatureÍ∞Ä 565Í∞úÏóêÏÑú 99Í∞úÎ°ú Ï§ÑÏñ¥ Ï∞®ÏõêÏù¥ Í∞êÏÜå ÎêêÏùåÏùÑ ÌôïÏù∏ Ìï† Ïàò ÏûàÎã§.

## 2. Data Modeling

```python
#inport libraries
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
```

### 2.1 ÍµêÏ∞®Í≤ÄÏ¶ù
- Î≥¥ÌÜµÏùÄ train set ÏúºÎ°ú Î™®Îç∏ÏùÑ ÌõàÎ†®, test setÏúºÎ°ú Î™®Îç∏ÏùÑ Í≤ÄÏ¶ùÌïúÎã§.Í≥†Ï†ïÎêú test setÏùÑ ÌÜµÌï¥ Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ Í≤ÄÏ¶ùÌïòÍ≥† ÏàòÏ†ïÌïòÎäî Í≥ºÏ†ïÏùÑ Î∞òÎ≥µÌïòÎ©¥, Í≤∞Íµ≠ ÎÇ¥Í∞Ä ÎßåÎì† Î™®Îç∏ÏùÄ test set ÏóêÎßå Ïûò ÎèôÏûëÌïòÎäî Î™®Îç∏Ïù¥ ÎêúÎã§.Ï¶â, test setÏóê Í≥ºÏ†ÅÌï©(overfitting)ÌïòÍ≤å ÎêòÎØÄÎ°ú, Îã§Î•∏ Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏ†∏ÏôÄ ÏòàÏ∏°ÏùÑ ÏàòÌñâÌïòÎ©¥ ÏóâÎßùÏù∏ Í≤∞Í≥ºÍ∞Ä ÎÇòÏôÄÎ≤ÑÎ¶¨Í≤å ÎêúÎã§.
- ÍµêÏ∞® Í≤ÄÏ¶ùÏùÄ train setÏùÑ train set + validation setÏúºÎ°ú Î∂ÑÎ¶¨Ìïú Îí§, validation setÏùÑ ÏÇ¨Ïö©Ìï¥ Í≤ÄÏ¶ùÌïòÎäî Î∞©ÏãùÏù¥Îã§.

```python
svc = SVC(C=1.0, kernel='rbf', gamma='scale')
scores = cross_val_score(svc, X_train_pca, y_train, cv=10, n_jobs=-1, scoring='accuracy') #cross_val_score(model,train,target,crossvalidation)
print('ÍµêÏ∞® Í≤ÄÏ¶ùÎ≥Ñ Ï†ïÌôïÎèÑ:',np.round(scores, 4))
print('ÌèâÍ∑† Í≤ÄÏ¶ù Ï†ïÌôïÎèÑ:', np.round(np.mean(scores), 4))
```

### 2.2 Î™®Îç∏ ÌïôÏäµ
#### SVM(kernalized Support Vector Machine)Ïù¥ÎûÄ?
- Ïª§ÎÑê ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†: ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Îã®ÏàúÌïú Ï¥àÌèâÎ©¥ hyperplaneÏúºÎ°ú Ï†ïÏùòÎêòÏßÄ ÏïäÎäî Îçî Î≥µÏû°Ìïú Î™®Îç∏ÏùÑ ÎßåÎì§ Ïàò ÏûàÎèÑÎ°ù ÌôïÏû•Ìïú Í≤ÉÏûÖÎãàÎã§
- Í≤∞Ï†ï Í≤ΩÍ≥Ñ(Decision Boundary), Ï¶â Î∂ÑÎ•òÎ•º ÏúÑÌïú Í∏∞Ï§Ä ÏÑ†ÏùÑ Ï†ïÏùòÌïòÎäî Î™®Îç∏
- SVMÏùÄ ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞(support vectors)Î•º ÏÇ¨Ïö©Ìï¥ÏÑú Í≤∞Ï†ï Í≤ΩÍ≥Ñ(Decision Boundary)Î•º Ï†ïÏùòÌïòÍ≥†, Î∂ÑÎ•òÎêòÏßÄ ÏïäÏùÄ Ï†êÏùÑ Ìï¥Îãπ Í≤∞Ï†ï Í≤ΩÍ≥ÑÏôÄ ÎπÑÍµêÌï¥ÏÑú Î∂ÑÎ•òÌïòÍ≤å ÎêúÎã§.
- Î∂ÑÎ•ò(Classification)ÎÇò ÌöåÍ∑Ä Î∂ÑÏÑù(Regression)Ïóê ÏÇ¨Ïö©Ïù¥ Í∞ÄÎä•ÌïòÎ©∞, ÌäπÌûà Î∂ÑÎ•òÏ™ΩÏóê ÏÑ±Îä• Îõ∞Ïñ¥ÎÇ®
- Í∏∞Ï°¥Ïùò ÌçºÏÖâÌä∏Î°†ÏùÄ Í∞ÄÏû• Îã®ÏàúÌïòÍ≥† Îπ†Î•∏ Î∂ÑÎ•ò Î™®ÌòïÏù¥ÏßÄÎßå Í≤∞Ï†ïÍ≤ΩÍ≥ÑÍ∞Ä Ïú†ÏùºÌïòÍ≤å Ï°¥Ïû¨ÌïòÏßÄ ÏïäÎäîÎã§.
- ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†(SVM)ÏùÄ ÌçºÏÖâÌä∏Î°† Í∏∞Î∞òÏùò Î™®ÌòïÏóê Í∞ÄÏû• ÏïàÏ†ïÏ†ÅÏù∏ Í≤∞Ï†ï Í≤ΩÍ≥ÑÎ•º Ï∞æÍ∏∞ ÏúÑÌï¥ Ï†úÌïú Ï°∞Í±¥ÏùÑ Ï∂îÍ∞ÄÌïú Î™®ÌòïÏù¥ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§.
- ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ : ÌÅ¥ÎûòÏä§ ÏÇ¨Ïù¥ Í≤ΩÍ≥ÑÏóê Í∞ÄÍπùÍ≤å ÏúÑÏπòÌïú Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ (Í≤∞Ï†ï Í≤ΩÍ≥ÑÏôÄ Ïù¥Îì§ ÏÇ¨Ïù¥Ïùò Í±∞Î¶¨Í∞Ä SVC Î™®Îç∏Ïùò dual_coef_Ïóê Ï†ÄÏû•ÎêúÎã§.)

#### Ïª§ÎÑê Í∏∞Î≤ïÏù¥ÎûÄ?
- Îç∞Ïù¥ÌÑ∞ÏÖãÏóê ÎπÑÏÑ†Ìòï ÌäπÏÑ±ÏùÑ Ï∂îÍ∞ÄÌïòÎ©¥ ÏÑ†Ìòï Î™®Îç∏ÏùÑ Îçî Í∞ïÎ†•ÌïòÍ≤å ÎßåÎì§ Ïàò ÏûàÏùå
- ÌïòÏßÄÎßå, Ïñ¥Îñ§ ÌäπÏÑ±ÏùÑ Ï∂îÍ∞ÄÌï¥Ïïº Ìï†ÏßÄ Ïïå Ïàò ÏóÜÍ≥†, ÌäπÏÑ±ÏùÑ ÎßéÏù¥ Ï∂îÍ∞ÄÌïòÎ©¥ Ïó∞ÏÇ∞ ÎπÑÏö©Ïù¥ Ïª§ÏßÑÎã§.
- Ïª§ÎÑê Í∏∞Î≤ï : ÏÉàÎ°úÏö¥ ÌäπÏÑ±ÏùÑ ÎßåÎì§ÏßÄ ÏïäÍ≥† Í≥†Ï∞®Ïõê Î∂ÑÎ•òÍ∏∞Î•º ÌïôÏäµÏãúÌÇ¨ Ïàò ÏûàÎèÑÎ°ù ÌïúÎã§. Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Í≥†Ï∞®ÏõêÏùò ÌäπÏßï Í≥µÍ∞ÑÏúºÎ°ú ÏÇ¨ÏÉÅÌï¥, ÏõêÎûòÏùò Ï∞®ÏõêÏóêÏÑ† Ìè¨Ïù¥ÏßÄ ÏïäÎçò ÏÑ†Ìòï(Ï¥àÌèâÎ©¥)Ïù¥ Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÎ•òÌï† Ïàò ÏûàÎèÑÎ°ù ÌïúÎã§.
- Í≥†Ï∞®Ïõê Í≥µÍ∞Ñ ÎßµÌïë Î∞©Î≤ï : Í∞ÄÏö∞ÏãúÏïà Ïª§ÎÑê, RBF (Radial Basis Function) Ïª§ÎÑê

```python
# Î™®Îç∏ ÌïôÏäµ
svc.fit(X_train_pca, y_train)
y_pred_svm=svc.predict(X_test_pca)
```

## 3. Data Evaluation
- TP(true positive): 1Î°ú ÏòàÏ∏°ÌñàÏùÑ Îïå Ï†ïÎãµÎèÑ 1Ïù∏ Í≤ΩÏö∞
- FP(false positive): 1Î°ú ÏòàÏ∏°ÌñàÏßÄÎßå Ï†ïÎãµÏùÄ 0Ïù∏ Í≤ΩÏö∞
- FN(false negative): 0ÏúºÎ°ú ÏòàÏ∏°ÌñàÏßÄÎßå Ï†ïÎãµÏùÄ 1Ïù∏ Í≤ΩÏö∞
- TN(true negative): 0ÏúºÎ°ú ÏòàÏ∏°ÌñàÏùÑ Îïå Ï†ïÎãµÎèÑ 0Ïù∏ Í≤ΩÏö∞
- recall(Ïû¨ÌòÑÏú®)=TP/(TP+FN): 
  ÏñëÏÑ± positive ÏºÄÏù¥Ïä§Îì§ÏóêÏÑú ÏßÑÏßú true ÏñëÏÑ± positiveÎ°ú ÏòàÏ∏°Ïóê ÏÑ±Í≥µÌïú ÌôïÎ•†
- precision(Ï†ïÎ∞ÄÎèÑ)=TP/(TP+FP): 
  ÏñëÏÑ± positiveÎùºÍ≥† ÌåêÎã®Ìïú Í≤ΩÏö∞ Ï§ëÏóê ÏßÑÏßú true ÏñëÏÑ± positiveÏù∏ ÌôïÎ•†
- F1_score=2‚àó(precision‚àórecall)/precision+recall

```python
# Import Libraries
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, recall_score, precision_score
```

```python
### print('confusion_matrix', confusion_matrix(y_pred_svm, y_test))
print('confusion_matrix', confusion_matrix(y_pred_svm, y_test))
print('accuracy score', accuracy_score(y_pred_svm, y_test))
print('f1_score', f1_score(y_test,y_pred_svm))
print('recall_score', recall_score(y_pred_svm, y_test))
print('precision_score',precision_score(y_test,y_pred_svm))
```
